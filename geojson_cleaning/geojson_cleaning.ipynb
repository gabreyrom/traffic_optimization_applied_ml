```python
import json
```

```python
with open("manh_geojson.txt", "r", encoding='utf-8') as f:
    infile = f.readline()
```

```python
len(infile)
```

```python
valid_ways = ["motorway", "trunk", "primary", "secondary", "tertiary", "unclassified", "residential",
             "motorway_link", "trunk_link", "primary_link", "secondary_link", "tertiary_link"]
```

```python
def index_way(highwaytag_index, file):
    start = file.rfind("}},{", 0, highwaytag_index) + 3        
    end = file.find("}},{", highwaytag_index) + 2
    
    return start, end
```

```python
def get_way_name(highwaytag_index, file):
    name_index = file.rfind('"name":', highwaytag_index-50, highwaytag_index) + 7
    name = file[name_index:highwaytag_index-1]
    return name
```

```python
def get_way_type(highwaytag_index, file):
    end_index = file.find("}", highwaytag_index)
    htype = file[highwaytag_index+11:end_index-1]
    return htype
```

```python
def trim_geojson(file):
    trimmed_file = ""
    trashed_file = ""
    index = file.find('"highway":')
    
    while index != -1:
        name = get_way_name(index, file)
        htype = get_way_type(index, file)
        way_start, way_end = index_way(index, file)
        
        if get_way_type(index, file) in valid_ways:
            trimmed_file = trimmed_file + file[way_start:way_end+1]
            print(f"saved way {name} of type {htype} at index {index}")
            # print(f"INDEX {index}\nWAY START {way_start}\nWAY END {way_end}\n",
                  # file[way_start:way_end+1], "\n")
        else:
            trashed_file = trashed_file + file[way_start:way_end+1]
            print(f"discarded {htype} at index {index}")
    
        old_index = index
        
        # go until there are no more '"highway":'
        index = file.find('"highway":', old_index+9)

    return trimmed_file, trashed_file
```

```python
trimmed, trashed = trim_geojson(infile)
```

```python
len(trimmed) + len(trashed)
```

```python
# formatting for trimmed file
if trimmed[0] != "{":
    trimmed = '{"' + trimmed # fill missing brackets
if trimmed[-1] == ",":
    trimmed = trimmed[:-1] + "]}" # remove final comma, close brackets
```

```python
# write to new txt file
with open("trimmed_manhattan.txt", "w") as trim_file:
    trim_file.write(trimmed)
    print("Wrote trimmed ways to txt file.")
```
